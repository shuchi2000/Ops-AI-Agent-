```python
import pandas as pd 

clients = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_clients.csv")
content = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_contents.csv")
deployment = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_deployments.csv")
evaluation = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_evaluations.csv")
programs  = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_programs.csv")
program_wise_learners = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_program_wise_learners.csv")
progress = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_progress.csv")
learners  = pd.read_csv(r"G:\My Drive\ML\2025\Project\Database Agent\data\enquriousV2_learners.csv")
```

## Let's prepare progress report

```python
programs.head()
```

```python
client_program = pd.merge(clients,programs,left_on='id', right_on='client_id',how='right')
client_program.shape
```

```python
client_program.head()
```

```python
client_program.rename({"id_y":"program_id","created_at_y":"Program_created_at"},axis =1 ,inplace=True)
client_program.drop(['id_x',"created_at_x"],axis=1,inplace=True)
```

```python
learners.head()
```

```python
learners.shape
```

```python
program_wise_learners.head()
```

```python
program_wise_learners.shape
```

```python
program_learner = pd.merge(program_wise_learners, learners, left_on='learner_id',right_on='id',how='left')
program_learner.shape
```

```python
program_learner.head()
```

```python
program_learner.drop(["id"],axis=1,inplace=True)
```

```python
client_program_learner = pd.merge(client_program,program_learner,on="program_id",how="right" )
client_program_learner.head()
```

```python
client_program_learner.rename(columns={'name_x':'Client_name','name_y':'learner_name'},inplace=True)
```

```python
client_program_learner.to_csv(r"G:\My Drive\ML\2025\Project\Database Agent\preparing_data\client_program_learner.csv")
```

```python
progress_deploy = pd.merge(progress,deployment, left_on="deployment_id", right_on="id", how="left")
progress_deploy.head()
```

```python
progress_deploy.rename(columns={'id_x':'Progress_id','created_at':"Progress_started_at"},inplace=True)
progress_deploy.drop(["id_y"],axis=1,inplace=True)
```

```python
progress_content_deploy = pd.merge(progress_deploy,content, left_on="content_id", right_on='id',how="left")
progress_content_deploy.head()
```

```python
progress_content_deploy.columns
```

```python
progress_content_deploy.rename(columns={ 'start_timestamp': 'deployment_start_timestamp', 'end_timestamp':'deployment_end_timestamp' ,'name':'content_name', 'type':'content_type', 'level':'content_level'}, inplace=True)
progress_content_deploy.drop(['id'],axis=1,inplace=True)
```

```python
progress_content_deploy.to_csv(r"G:\My Drive\ML\2025\Project\Database Agent\preparing_data\progress_content_deploy.csv")
```

```python
progress_df = pd.merge(client_program_learner,progress_content_deploy,on=['program_id','learner_id'],how='right')
```
